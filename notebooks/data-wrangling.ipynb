```python
import pandas as pd
from datetime import datetime
```

```python
file = '../data/lista-premieres.csv'
df = pd.read_csv(file, encoding='latin-1')
```

```python
df.head()
```

```python
time_mask = '%d-%b-%Y'
```

```python
df['Fecha de nacimiento'] = df['Fecha de nacimiento'].apply(lambda x: x[:-2] + '19' +x[-2:])
df['Fecha de nacimiento'] = df['Fecha de nacimiento'].apply(lambda x: datetime.strptime(x, time_mask))
```

```python
df['Fecha de juramentacion'] = df['Fecha de juramentacion'].apply(lambda x: x[:-2] + '20' +x[-2:])
df['Fecha de juramentacion'] = df['Fecha de juramentacion'].apply(lambda x: datetime.strptime(x, time_mask))
```

```python
df.head()
```

```python
df['Edad al juramentar'] = df['Fecha de juramentacion'] - df['Fecha de nacimiento']
df['Edad al juramentar'] = df['Edad al juramentar'].apply(lambda x: int(x.days / 365.25))
df.head()
```

```python
df['Fecha final'] = df['Fecha de juramentacion'].shift()
df['Fecha final'].iloc[0] = datetime.strptime('26-Aug-2021', time_mask)
```

```python
df.head()
```

```python
df['Meses en el cargo'] = df['Fecha final'] - df['Fecha de juramentacion']
df['Meses en el cargo'] = df['Meses en el cargo'].apply(lambda x: x.days / 30)
df.head()
```

```python
df['Nacida/o en Lima?'] = 'Fuera de Lima'
df.loc[df['Departamento']=='Lima', 'Nacida/o en Lima?'] = 'En Lima'
```

```python
df.head()
```

```python
df
```

```python
pc = df['Nombre'].iloc[0]
fs = df['Nombre'].iloc[1]
mm = df['Nombre'].iloc[2]
mv = list(df['Nombre'].iloc[3:8])
ppk = list(df['Nombre'].iloc[8:10])
oh = list(df['Nombre'].iloc[10:17])
ag = list(df['Nombre'].iloc[17:22])
at = list(df['Nombre'].iloc[22:27])
vp = df['Nombre'].iloc[27]
```

```python
df['Presidente'] = ''
df.loc[df['Nombre']==pc, 'Presidente'] = 'PC'
df.loc[df['Nombre']==fs, 'Presidente'] = 'FS'
df.loc[df['Nombre']==mm, 'Presidente'] = 'MM'
df.loc[df['Nombre'].isin(mv), 'Presidente'] = 'MV'
df.loc[df['Nombre'].isin(ppk), 'Presidente'] = 'PPK'
df.loc[df['Nombre'].isin(oh), 'Presidente'] = 'OH'
df.loc[df['Nombre'].isin(ag), 'Presidente'] = 'AG'
df.loc[df['Nombre'].isin(at), 'Presidente'] = 'AT'
df.loc[df['Nombre']==vp, 'Presidente'] = 'VP'
df['Presidente'].iloc[4] = 'MV'
df['Presidente'].iloc[7] = 'MV'
```

```python
df['Nombre2'] = df['Nombre'] + ' (' + df['Presidente'] + ')'
```
